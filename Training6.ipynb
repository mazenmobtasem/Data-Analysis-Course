### SAT => Scholastic Assessment Test

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

f = 'SAT_GPA.csv'

table = pd.read_csv(f)
```

```python
table.head()
```

```python
X = table['SAT'].values
Y = table['GPA'].values
```

```python
type(X)
```

```python
len(X)
```

```python
plt.scatter(X,Y);
```

```python
plt.boxplot(Y, showfliers = False);
```

```python
from IPython.display import Image
Image('SLR.PNG')
```

```python
## your code here
a = sum([(x - np.mean(X)) * (y - np.mean(Y)) for x,y in zip(X,Y)]) / sum( [(x - np.mean(X))**2 for x in X])
b = np.mean(Y) - a * np.mean(X)
```

```python
a
```

```python
b
```

```python
yhat = lambda x: a * x + b
```

```python
plt.scatter(X,Y)
plt.plot([X.min(), X.max()], [yhat(X.min()), yhat(X.max())], color = 'g')
plt.scatter(X.mean(), Y.mean(), marker = 'D', c = 'r')
```

```python
Image('R2.PNG')
```

```python
Yhat = np.array([yhat(x) for x in X])
```

```python
## your code here
TSS = sum([(y - np.mean(Y))**2 for y in Y])
RSS = sum([(y - yhat(x))**2 for x,y in zip(X,Y)])
R2 = 1 - RSS/TSS
```

```python
print(f'TSS is: {TSS}, RSS is: {RSS}, R2 is: {R2}')
```
